### Actividad
Obtenga el último trabajo de Álvaro Uribe Vélez

````python
persona[persona['Nombre'].str.lower().str.contains('álvaro uribe vélez')].iloc[0]['id']
#contains es lo suficientemente flexible como para permitir búsquedas de frases parcialmente.
#Aquí básicamente se hizo un filtro mediante una máscara. Extrajo la fila tal que el valor correspondiente a la columna
#'Nombre' en minúsculas sea 'álvaro uribe vélez'.
#Si en una búsqueda hay hartos Álvaro Uribe Vélez, con iloc puedo extraer el último o el primer elemento dentro de esos
#resultados de búsqueda. 
#Luego a esa fila le sacó el valor correspondiente a la columna 'id'.
````

````python
cc=persona[persona['Nombre'].str.lower().str.contains('álvaro uribe vélez')].iloc[0].get('id') #Identificador.
#Mejor que ['id'] porque si no hay tal columna, no retornará error sino un resultado vacío.
prof=trabajo[trabajo['id']==cc]
prof=prof[prof.Inicio==prof.Inicio.max()]
prof=prof.iloc[0].get('Cargo')
prof
````

````python
cc=persona[persona['Nombre'].str.lower().str.contains('álvaro uribe vélez')].iloc[0].get('id')
trabajo[trabajo['id']==cc]['Cargo'].iloc[-1]
````
